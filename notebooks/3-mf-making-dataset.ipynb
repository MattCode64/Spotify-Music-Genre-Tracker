```python
import pandas as pd
```

```python
# Load data
df = pd.read_csv('../data/preprocessed/preprocessed_dataset.csv')
```

```python
df.head()
```

```python
col_to_standardize = ['popularity', 'duration_ms', 'explicit', 'danceability', 'energy', 'loudness', 'speechiness',
                      'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
```

```python
# Standardize with sklearn
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[col_to_standardize] = scaler.fit_transform(df[col_to_standardize])
```

```python
df.describe()
```

```python
# Save the dataset
df.to_csv('../data/processed/dataset_ready.csv', index=False)
```
